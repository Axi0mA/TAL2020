```python
import wikipedia
import spacy
import networkx as nx
import matplotlib.pyplot as plt

wikipedia.set_lang("es")
nlp = spacy.load('es_core_news_lg') 
plt.rcParams['figure.figsize'] = [150, 150]
```

```python
def obtener_nombres(personaje):
    page = wikipedia.page(personaje)
    sents = nlp(page.content) 
    entities={}

    for ent in sents.ents:   
        # We check if entity is a Person type using the SpaCy model
        if(ent.label_=="PER"):
            #We tokenize the entity
            tokenized_entity=(ent.text).split(" ") 

            # We preserve only the entities that has between 2 abd 4 tokens (Usual name annotation in Chile)
            if ((len(tokenized_entity)>1) and len(tokenized_entity)<=6):

                entity_full_name = ent.text.split(",")[0].split(".")[0]

                if entity_full_name in entities:
                    entities[entity_full_name] += 1

                else:
                    entities[entity_full_name] = 1
    return page, entities
```

```python
def imprimir(G, nx):
    elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 2]
    esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] <= 2]

    pos = nx.spring_layout(G)  # positions for all nodes

    # nodes
    nx.draw_networkx_nodes(G, pos, node_size=200)

    # edges
    nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6)
    nx.draw_networkx_edges(
        G, pos, edgelist=esmall, width=6, alpha=0.5, edge_color="b", style="dashed"
    )

    # labels
    nx.draw_networkx_labels(G, pos, font_size=30, font_family="sans-serif")
    plt.savefig('foo.png', format="PNG")
    plt.axis("off")
    plt.figure(figsize=(1,1), dpi= 20, facecolor='w', edgecolor='k')
    
    #fig.show()
```

```python
def añadir_entidad(nombre):
    lista=[]
    page, entities = obtener_nombres(nombre)
    #Añadir nodo raíz
    G.add_node(page.title)
    for key,value in entities.items():
        #Añadir nodos
        G.add_node(key, weight = len(entities.items()))
        #Añadir aristas
        G.add_edge(page.title, key, weight = value)
        lista.append(key)
    return lista
```

```python
def extraer_info(lista):
    for i in lista:
        try:
            lista_ = añadir_entidad(i)
            retorno = [j for j in lista_]
        except:
            print(i, "no encontrado")
        
    return retorno
```

```python
def iteraciones(semilla, n):
    lista = [semilla]
    for i in range (n):
        lista = extraer_info(lista)
```
